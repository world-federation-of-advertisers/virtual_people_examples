Geometric Shredding is a technique to model multiple users sharing a device and therefore sharing a user identifier.

Geometic Shred is to be generated from user identifier $c$ and event identifier $i$ as follows. Given a _shredding probability_ parameter $\psi$, which corresponds to the success probability parameter of geometric distribution as $p = 1 - \psi$, we first generate a geometrically distributed hash $t$ of event identifier $i$ and then shred $s$ is simply computed as a hash of a pair $(c, t)$.

Using a formula maps that exponentially distributed hash to geometrically distributed hash we can generate shred with a uniform float hash function $f$ and a binary integer hash function $h$ as follows:

$$s = h\left(c, \left\lfloor - \log\left(\frac{f(i)}{-\log(\psi)}\right) \right\rfloor \right)  $$

Next we run a study, where we generate synthetic campaign data that simulates universe in which device sharing occurs. In this universe $cookies \rightarrow people$ curve has $\kappa > 1$. Our study illustrates that picking an appropriate $\psi$ for geometric shredding we can create $\psi$-shreds such that $\psi {\text -} shreds \rightarrow people$ curve has  $\kappa < 1$.

Other forms of shredding may be considered in the future to minimize noise under specific practical cirtumstances.


# Synthetic Logs Generation

We synthesize logs with events of the form `(event_id, person_id, cookie, campaign, shred)`.

In these logs

* **cookie sharing** is simulated by each cookie being shared between 1, 2 or 3 people,
* **non-uniform ADF** is simulated by people with even identifiers being twice as active as people with odd identifiers.

```python
%%logica Activity, Reach, SaveSyntheticLogs


SyntheticLogsConfig(
    population: 1000,
    num_visits: 15000,
    cookie_population: 3000,
    max_multiplicity: 10
);

# Positive hash.
Fingerprint(x) = Abs(FarmFingerprint(x));

# Float hash between 0 and 1.
FloatFingerprint(x) = Fingerprint(x) / 9223372036854775807;

@Ground(Activity);
Activity(event_id:,
         person_id:,
         cookie:,
         campaign:,
         shred:) distinct :-
  SyntheticLogsConfig(
      population:,
      num_visits:,
      cookie_population:,
      max_multiplicity:
  ),
  idx in Range(num_visits),
  multiplicity == Fingerprint("multiplicity:" ++ ToString(idx)) % max_multiplicity,
  j in Range(multiplicity),
  cookie == Fingerprint("cookie:" ++ ToString(idx)) % cookie_population,
  num_people == Fingerprint("num_people:" ++ ToString(idx)) % 3 + 1,
  i in Range(num_people),
  person_id ==
    2 * (Fingerprint("person-c1:" ++ ToString(cookie) ++ "-" ++ ToString(i)) % Div(population, 2)) +
    (Fingerprint("person-c2:" ++ ToString(cookie) ++ "-" ++ ToString(i)) % 3) % 2,
  # Campaign is sampled in such a way that we get campaigns of different
  # sizes, so the curve is clearly visible from the scatterchart.
  h == FloatFingerprint("h:" ++ ToString(idx)),
  campaign == ToInt64(h ^ 0.2 * 20),
  # EventId is just a hash of all parameters.
  event_id == Fingerprint(
      "event:" ++ ToString(person_id) ++ "-" ++ ToString(cookie) ++ "-" ++
      ToString(campaign) ++ "-" ++ ToString(j)),
  # True shred is a hash of the user and a cookie that they use.
  shred == Fingerprint("shred:" ++ ToString(person_id) ++ "-" ++ ToString(cookie));


@OrderBy(Reach, "campaign");
Reach(campaign:,
      cookies? Count= cookie, people? Count= person_id,
      shreds? Count= shred,
      impressions? += 1) distinct :-
  Activity(person_id:, cookie:, campaign:, shred:);

@Ground(SyntheticLogs);
SyntheticLogs(event_id:, campaign:, cookie:, person_id:) :-
  Activity(person_id:, cookie:, campaign:, event_id:);

SaveSyntheticLogs() += 1 :- SyntheticLogs();
```

# Purposeful shredding

Next we create traffic data where for each campaign we count people and shreds, running through a grid of values for $\psi$.

```python
%%logica TestGeometic, PsiToTry, EstimatedShredReach

# Positive hash.
Fingerprint(x) = Abs(FarmFingerprint(x));
# Float hash between 0 and 1.
FloatFingerprint(x) = Fingerprint(x) / 9223372036854775807;
# Exponentially distributed fingerprint.
ExpFingerprint(x) = - Log(FloatFingerprint(x));

PsiToTry() = 0.04 * i :- i in Range(9);

TestGeometic(expected: 0.35,
             observed:) :-
  observed Avg= (ToInt64(s == Fingerprint("10-shred-0")) :-
                 s == Shred(10, i, 0.65),  # 0.65 shredding probability
                 i in Range(10000));


Shred(cookie, event_id, parameter) = Fingerprint(cookie ++ "-shred-" ++ shred_hash) :-
  shred_hash == (
      if parameter == 0 then
        0  # No shredding.
      else if parameter == 1 then
        event_id  # Certain shredding.
      else
        ToInt64(Floor(ExpFingerprint(ToString(event_id)) / d))
  ),
  d == - Log(parameter);

@Ground(SyntheticLogs);  # Read SyntheticLogs.

EstimatedShredReach(campaign:,
                    psi:,
                    cookies? Count= cookie,
                    people? Count= person_id,
                    shreds? Count= shred_id,
                    events? Count= event_id) distinct :-
  SyntheticLogs(event_id:, campaign:, cookie:, person_id:),
  psi == PsiToTry(),
  shred_id == Shred(cookie, event_id, psi);
```

```python
pyplot.figure()
pyplot.title("Cookies vs People")
pyplot.scatter(EstimatedShredReach['cookies'],
               EstimatedShredReach['people'])
pyplot.plot([0, 1000], [0, 1000], linestyle='--')
pyplot.figure()
pyplot.title("Cookies vs Purposeful Shreds")
pyplot.scatter(EstimatedShredReach['shreds'],
               EstimatedShredReach['people'],
               color=(['red', 'blue', 'green', 'black', 'olive', 'cyan',
                       'violet', 'brown', 'navy']) * (
                   len(EstimatedShredReach) // 9))
pyplot.plot([0, 1000], [0, 1000], linestyle='--')


None
```

# Library for ADM training

```python
from scipy import optimize
import numpy
from matplotlib import pyplot


def fit_alphas(ds, signal, target):
  c = (1 - numpy.exp(- ds.dot(signal.T))).T
  result, _ = optimize.nnls(c, target, maxiter=1000)
  return result

class DiracMixture(object):
  def __init__(self, alphas, ds, scale=1.0):
    self.alphas = alphas
    self.ds = ds
    self.scale = scale

  def __call__(self, signal):
    return numpy.hstack(self.alphas).dot((1 - numpy.exp(- self.ds.dot(signal.T / self.scale)))) * self.scale

  def __str__(self):
    return (
        'DiracMixture(\n    ' +
        ' + \n    '.join(
            '%.3f * %s' % (alpha, delta)
            for alpha, delta in zip(adm.alphas, adm.ds)) +
        ',\n    scale=%.3f)' % self.scale)

def sampled_from_gaussian_mixture(centers, center_probabilities,
                                  num_new_centers, sigma):
  dim = centers.shape[1]
  new_centers = centers[numpy.random.choice(
      len(center_probabilities),
      num_new_centers,
      replace=True, p=center_probabilities)]
  noise = numpy.random.normal(0, sigma, (num_new_centers, dim))
  new_centers += noise
  min_coord = new_centers.min(axis=1)
  new_centers = new_centers[(min_coord > 0)]
  return new_centers

def fit_adaptive_dirac_mixture(signal, target,
                               new_centers_at_each_step=10,
                               new_centers_sigma=0.01,
                               max_steps=1000,
                               initial_centers=None,
                               iteration_callback=None):
  dim = signal.shape[1]
  centers = initial_centers or numpy.vstack([numpy.ones(dim)])
  alphas = numpy.ones(len(centers)) / len(centers)
  for i in range(max_steps):
    center_probabilities = alphas / sum(alphas)
    new_centers = sampled_from_gaussian_mixture(
        centers, center_probabilities,
        new_centers_at_each_step, new_centers_sigma)
    centers = numpy.vstack([centers, new_centers])
    alphas = fit_alphas(centers,
                        signal,
                        target)
    centers = centers[alphas > 0]
    alphas = alphas[alphas > 0]
    if iteration_callback:
      iteration_callback(i, alphas, centers)
  return alphas, centers

class AdaptiveDiracMixture(DiracMixture):
  def __init__(self, scale=1.0):
    self.alphas = None
    self.ds = None
    self.scale = scale

  def fit(self, signal, target,
          new_centers_at_each_step=10,
          new_centers_sigma=0.01,
          max_steps=1000,
          initial_centers=None,
          iteration_callback=None):
    self.alphas, self.ds = fit_adaptive_dirac_mixture(
        signal / self.scale, target / self.scale,
        new_centers_at_each_step=10,
        new_centers_sigma=0.01,
        max_steps=1000,
        initial_centers=None,
        iteration_callback=None)

toy_signal = numpy.random.uniform(0, 1, (1000, 2))
toy_ds = [[1.5, 0.5],
          [0.5, 1.5]]
toy_alphas = numpy.ones(len(toy_ds)) / len(toy_ds)
toy_dm = DiracMixture(toy_alphas, numpy.array(toy_ds))
toy_target = toy_dm(toy_signal)

adm = AdaptiveDiracMixture()
adm.fit(toy_signal, toy_target, max_steps=10)

print('Toy Dirac Mixture:')
print(adm)
```

# Fitting Shred-based Reach Curves

Finally for each $\psi$ we fit a reach curves mapping $\psi$-shreds $\rightarrow$ people.

We observe that as expected $\kappa$ value decreases with the increase of $\psi$

While it's not the case in our simplistic simulation, we know that in practice
higher levels of shredding would have higher noise of the shreds vs people scatterchart. Thus we should chose the smallest $\psi$ which gives $\kappa \leq 1$.

```python
import numpy

pyplot.figure(figsize=(20, 10))
seen_kappa_less_than_1 = False
for idx, psi in enumerate(PsiToTry.logica_value):
  ax = pyplot.subplot(3, 3, idx + 1)
  df = EstimatedShredReach[
      EstimatedShredReach['psi'] == psi]

  ax.set_title("Shredding probability: %.3f" % psi +
               ("\nNOTE: fit is so good because simulation "
                "isn't reflecting noise of impressions"
                if psi == 1.0 else ""))
  pyplot.xlabel('shreds')
  pyplot.ylabel('people')
  pyplot.scatter(df['shreds'], df['people'])
  pyplot.axhline(1000, linestyle=':')
  xs = numpy.linspace(0, max(df['shreds']))
  pyplot.plot([0, 1000], [0, 1000], linestyle='--', label='y = x')

  # Adding (100k, 1k) points to softly enforce correct limit of the curve.
  signal = numpy.vstack((df[['shreds']], [[1000000]] * 10))
  target = numpy.hstack((df['people'], [1000] * 10))
  adm = AdaptiveDiracMixture(scale=1000)
  adm.fit(signal, target)
  estimates = adm(xs.reshape((len(xs), 1)))
  kappa = numpy.dot(adm.alphas, adm.ds[:, 0])
  pyplot.plot(xs, estimates, label='adm with kappa=%.3f' % kappa)
  pyplot.legend(loc='best')
  if not seen_kappa_less_than_1 and kappa < 1:
    seen_kappa_less_than_1 = True
    print('ADM for psi=%.3f:' % psi)
    print(adm)

pyplot.subplots_adjust(left=0, bottom=0, right=1, top=1,
                       wspace=0.15, hspace=0.3)
```
